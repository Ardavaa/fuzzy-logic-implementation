```python
import pandas as pd
import numpy as np

# Fungsi Keanggotaan untuk Tekanan Darah (trestbps)
def fuzzification_trestbps(x):
    if x < 90:
        return [1, 0, 0]  # Rendah: 1, Normal: 0, Tinggi: 0
    elif 90 <= x <= 130:
        return [(130 - x) / 40, (x - 90) / 40, 0]  # Menurun dari rendah ke normal
    elif 130 < x <= 170:
        return [0, (170 - x) / 40, (x - 130) / 40]  # Menurun dari normal ke tinggi
    else:
        return [0, 0, 1]  # Tinggi: 1

# Fungsi Keanggotaan untuk Kolesterol (chol)
def fuzzification_chol(x):
    if x < 200:
        return [1, 0, 0]  # Rendah: 1, Sedang: 0, Tinggi: 0
    elif 200 <= x <= 240:
        return [(240 - x) / 40, (x - 200) / 40, 0]  # Menurun dari rendah ke sedang
    elif 240 < x <= 280:
        return [0, (280 - x) / 40, (x - 240) / 40]  # Menurun dari sedang ke tinggi
    else:
        return [0, 0, 1]  # Tinggi: 1

# Fungsi Keanggotaan untuk Usia (age)
def fuzzification_age(x):
    if x < 30:
        return [1, 0, 0]  # Muda: 1, Dewasa: 0, Tua: 0
    elif 30 <= x <= 50:
        return [(50 - x) / 20, (x - 30) / 20, 0]  # Menurun dari muda ke dewasa
    elif 50 < x <= 70:
        return [0, (70 - x) / 20, (x - 50) / 20]  # Menurun dari dewasa ke tua
    else:
        return [0, 0, 1]  # Tua: 1
```

```python
# Fungsi Inferensi Mamdani
def inferensi_mamdani(trestbps_fuzz, chol_fuzz, age_fuzz):
    # 27 aturan untuk kombinasi keanggotaan
    aturan = []

    # Kombinasi untuk Rendah pada trestbps, Rendah pada chol, Muda pada age
    aturan.append(min(trestbps_fuzz[0], chol_fuzz[0], age_fuzz[0]))  # Rendah - Rendah - Muda -> Output: Rendah
    aturan.append(min(trestbps_fuzz[0], chol_fuzz[0], age_fuzz[1]))  # Rendah - Rendah - Dewasa -> Output: Sedang
    aturan.append(min(trestbps_fuzz[0], chol_fuzz[0], age_fuzz[2]))  # Rendah - Rendah - Tua -> Output: Tinggi

    # Kombinasi untuk Rendah pada trestbps, Rendah pada chol, Dewasa pada age
    aturan.append(min(trestbps_fuzz[0], chol_fuzz[0], age_fuzz[0]))  # Rendah - Rendah - Muda -> Output: Rendah
    aturan.append(min(trestbps_fuzz[0], chol_fuzz[0], age_fuzz[1]))  # Rendah - Rendah - Dewasa -> Output: Sedang
    aturan.append(min(trestbps_fuzz[0], chol_fuzz[0], age_fuzz[2]))  # Rendah - Rendah - Tua -> Output: Tinggi

    # Kombinasi untuk Rendah pada trestbps, Rendah pada chol, Tua pada age
    aturan.append(min(trestbps_fuzz[0], chol_fuzz[0], age_fuzz[0]))  # Rendah - Rendah - Muda -> Output: Rendah
    aturan.append(min(trestbps_fuzz[0], chol_fuzz[0], age_fuzz[1]))  # Rendah - Rendah - Dewasa -> Output: Sedang
    aturan.append(min(trestbps_fuzz[0], chol_fuzz[0], age_fuzz[2]))  # Rendah - Rendah - Tua -> Output: Tinggi

    # Kombinasi untuk Normal pada trestbps, Rendah pada chol, Muda pada age
    aturan.append(min(trestbps_fuzz[1], chol_fuzz[0], age_fuzz[0]))  # Normal - Rendah - Muda -> Output: Rendah
    aturan.append(min(trestbps_fuzz[1], chol_fuzz[0], age_fuzz[1]))  # Normal - Rendah - Dewasa -> Output: Sedang
    aturan.append(min(trestbps_fuzz[1], chol_fuzz[0], age_fuzz[2]))  # Normal - Rendah - Tua -> Output: Tinggi

    # Kombinasi untuk Normal pada trestbps, Rendah pada chol, Dewasa pada age
    aturan.append(min(trestbps_fuzz[1], chol_fuzz[0], age_fuzz[0]))  # Normal - Rendah - Muda -> Output: Rendah
    aturan.append(min(trestbps_fuzz[1], chol_fuzz[0], age_fuzz[1]))  # Normal - Rendah - Dewasa -> Output: Sedang
    aturan.append(min(trestbps_fuzz[1], chol_fuzz[0], age_fuzz[2]))  # Normal - Rendah - Tua -> Output: Tinggi

    # Kombinasi untuk Normal pada trestbps, Rendah pada chol, Tua pada age
    aturan.append(min(trestbps_fuzz[1], chol_fuzz[0], age_fuzz[0]))  # Normal - Rendah - Muda -> Output: Rendah
    aturan.append(min(trestbps_fuzz[1], chol_fuzz[0], age_fuzz[1]))  # Normal - Rendah - Dewasa -> Output: Sedang
    aturan.append(min(trestbps_fuzz[1], chol_fuzz[0], age_fuzz[2]))  # Normal - Rendah - Tua -> Output: Tinggi

    # Kombinasi untuk Tinggi pada trestbps, Rendah pada chol, Muda pada age
    aturan.append(min(trestbps_fuzz[2], chol_fuzz[0], age_fuzz[0]))  # Tinggi - Rendah - Muda -> Output: Rendah
    aturan.append(min(trestbps_fuzz[2], chol_fuzz[0], age_fuzz[1]))  # Tinggi - Rendah - Dewasa -> Output: Sedang
    aturan.append(min(trestbps_fuzz[2], chol_fuzz[0], age_fuzz[2]))  # Tinggi - Rendah - Tua -> Output: Tinggi

    # Kombinasi untuk Tinggi pada trestbps, Rendah pada chol, Dewasa pada age
    aturan.append(min(trestbps_fuzz[2], chol_fuzz[0], age_fuzz[0]))  # Tinggi - Rendah - Muda -> Output: Rendah
    aturan.append(min(trestbps_fuzz[2], chol_fuzz[0], age_fuzz[1]))  # Tinggi - Rendah - Dewasa -> Output: Sedang
    aturan.append(min(trestbps_fuzz[2], chol_fuzz[0], age_fuzz[2]))  # Tinggi - Rendah - Tua -> Output: Tinggi

    # Kombinasi untuk Tinggi pada trestbps, Rendah pada chol, Tua pada age
    aturan.append(min(trestbps_fuzz[2], chol_fuzz[0], age_fuzz[0]))  # Tinggi - Rendah - Muda -> Output: Rendah
    aturan.append(min(trestbps_fuzz[2], chol_fuzz[0], age_fuzz[1]))  # Tinggi - Rendah - Dewasa -> Output: Sedang
    aturan.append(min(trestbps_fuzz[2], chol_fuzz[0], age_fuzz[2]))  # Tinggi - Rendah - Tua -> Output: Tinggi

    return aturan

# Defuzzification menggunakan metode Centroid
def defuzzification(aturan):
    return sum(aturan) / len(aturan) if len(aturan) != 0 else 0

# Proses untuk Fuzzification, Inferensi, dan Defuzzification
id = range(100)  # Misalnya, data ID
trestbps = np.random.randint(80, 180, size=100)  # Data tekanan darah (random)
chol = np.random.randint(150, 300, size=100)  # Data kolesterol (random)
age = np.random.randint(20, 80, size=100)  # Data usia (random)

y = []

for x in range(len(trestbps)):
    # Fuzzifikasi untuk 'trestbps', 'chol', dan 'age'
    trestbps_fuzz = fuzzification_trestbps(trestbps[x])
    chol_fuzz = fuzzification_chol(chol[x])
    age_fuzz = fuzzification_age(age[x])
    
    # Inferensi Mamdani untuk menghasilkan aturan
    aturan = inferensi_mamdani(trestbps_fuzz, chol_fuzz, age_fuzz)
    
    # Defuzzifikasi untuk menghasilkan nilai kelayakan
    kelayakan = defuzzification(aturan)
    
    # Menyimpan hasil untuk evaluasi
    y.append(kelayakan)

# Menampilkan hasil evaluasi
df = pd.DataFrame({
    'ID': id,
    'Trestbps': trestbps,
    'Chol': chol,
    'Age': age,
    'Kelayakan': y
})

print(df.head())
```
